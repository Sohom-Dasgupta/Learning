# Polynomial Regression

## Importing the libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```
dataset=pd.read_csv('Position_Salaries.csv')
x=dataset.iloc[:,1:-1].values
#here we took the columns from index 1 because the first column data needs encoding but the second column values are distinct for every row in 1st column. So for every pos, level is different and hence we remove pos.
y=dataset.iloc[:,-1].values
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin_regressor=LinearRegression()
lin_regressor.fit(x,y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4) #poly_reg is the object where we define feature matrix will be of degree 2
x_poly=poly_reg.fit_transform(x) #x_poly will hold the new matrix with 2 columns of x and x^2
poly_lin_reg=LinearRegression() #now intialize the linear reg model
poly_lin_reg.fit(x_poly,y)# train the model with new feature matrix
```

## Visualising the Linear Regression results

```
plt.scatter(x,y,color='red')
plt.plot(x,lin_regressor.predict(x),color='blue')
plt.title('Truth or Bluff by LIN REG')
plt.xlabel("position level")
plt.ylabel("Salaries")
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(x,y,color='red')
plt.plot(x,poly_lin_reg.predict(x_poly),color='blue')
plt.title('Truth or Bluff by POLY LIN REG')
plt.xlabel("position level")
plt.ylabel("Salaries")
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, poly_lin_reg.predict(poly_reg.fit_transform(x_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```
lin_regressor.predict([[6.5]])
 #the parameter for the predict is an array which is 2d in nature [[6.5]] means 1 row and 1 column with 6.5. [[6.5,2.5],[2,3]] means 2 rows with 2 columns
```
